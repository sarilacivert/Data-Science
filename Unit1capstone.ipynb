```python
import json
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
sample_json_df = pd.read_json('/Users/lacivert/2014_2017.json')
```

```python
#assign DataFrame
df = pd.DataFrame(sample_json_df)
```

```python
#list column names
list(df.columns.values)
```

```python
one_month, six_month, one_year, ten_year, thirty_year  = df.loc[:,'1 mo'], df.loc[:,'6 mo'], df.loc[:,'1 yr'], df.loc[:,'10 yr'], df.loc[:,'30 yr']
```

```python
np.mean(one_month),np.mean(six_month),np.mean(one_year),np.mean(ten_year),np.mean(thirty_year)
```

```python
sns.distplot(one_month,label='1 Month')
sns.distplot(six_month,label='6 Month')
sns.distplot(one_year, label='1 Year')
plt.xlabel('1 Month, 6 Month and 1 Year Maturities',color='green')
plt.ylabel('Number of Observations',color='green')
plt.title('Daily Distplot 2014 - 2017', color='green')
plt.legend()
```

```python
sns.distplot(ten_year,label='10 Year')
sns.distplot(thirty_year,label='30 Year')
plt.xlabel('10 Year and Thirty Year Maturities',color='blue')
plt.ylabel('Number of Observations',color='blue')
plt.title('Daily Distplot 2014 - 2017', color='blue')
plt.legend()
```

```python
import scipy.stats as stats
```

```python
t_1mo_6mo, p_1mo_6mo = stats.ttest_ind(one_month,six_month)
print (t_1mo_6mo, p_1mo_6mo)
```

```python
t_1mo_1yr, p_1mo_1yr = stats.ttest_ind(one_month,one_year)
print (t_1mo_1yr, p_1mo_1yr)
```

```python
t_6mo_1yr, p_6mo_1yr = stats.ttest_ind(six_month,one_year)
print (t_6mo_1yr, p_6mo_1yr)
```

```python
t_10yr_30yr, p_10yr_30yr = stats.ttest_ind(ten_year,thirty_year)
print (t_10yr_30yr, p_10yr_30yr)
```
