```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
x = np.random.normal(size=100)
```

```python
sns.kdeplot(x, shade=True, color='orange')
plt.title('Kernal Density Estimate = kdeplot', color='green')
plt.xlabel('x random variable with size 100',color='green')
plt.ylabel('Density', color='green')
```

```python
sns.distplot(x, color='blue', vertical=True)
plt.title('mathplotlib hist, seaborn kdeplot, seaborn rugplot is this displot', color='green')
plt.ylabel('x random variable with size 100',color='green')
plt.xlabel('Density', color='green')
```

```python
sns.rugplot(x,height=0.5)
plt.title('rugplot',color='green')
plt.xlabel('x random variable with size 100',color='green')
plt.ylabel('Density', color='green')
```

```python
sns.boxplot(x),plt.title('Boxplot',color='green')
plt.xlabel('abscissas',color='green')
plt.ylabel('ordinates', color='green')
```

```python
#bivariate 
j, k = np.random.normal(size=100), np.random.normal(size=100)
```

```python
df = pd.DataFrame(j,k)
```

```python
sns.jointplot(j,k,data=df), plt.title('jointplot',color='green')
```

```python
sns.jointplot(j,k,data=df,kind='kde'), plt.title('jointplot kde',color='blue')
```

```python
sns.jointplot(j,k,data=df,kind='hex',color='magenta'), plt.title('jointplot kde',color='magenta')
```
